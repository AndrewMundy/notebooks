# The Effects of FFT on Exponentiation using Circular Convolution

First we'll import some useful utilities:

```python
%cd ../hrr-playground/
from Holographic import *
import numpy as np

%pylab inline
from pylab import *
```

We can now generate a sample vector of 512 dimensions: (Item 1: use powers of 2)

```python
a = vec_generate( 512 )
```

Converting this to the Frequency Domain using FFT:

```python
fft_a = np.fft.fft( a )

# Plotting the magnitude of the vector in frequency domain
figure()
plot( arange( fft_a.size ), abs( fft_a ) )
xlim( (0, fft_a.size - 1) )
xlabel( "Component" )
ylabel( "Magnitude" )
title( "Magnitude of the Frequency Representation of the Vector" )
show()

# Exponentiating
fft_a2 = fft_a ** 2
a2 = real( np.fft.ifft( fft_a2 ) )

vec_magnitude = lambda v : sqrt( sum( array( v ) ** 2 ) )
print "||a||   = % 5.3f" % vec_magnitude( a )
print "||a^2|| = % 5.3f" % vec_magnitude( a2 )
```

Repeating this process for the $\delta$ function of the same dimensionality:

```python
delta = hstack( [1, zeros( 511 ) ] )
fft_delta = np.fft.fft( delta )

# Plotting the magnitude of the vector in frequency domain
figure()
plot( arange( fft_delta.size ), abs( fft_delta ) )
xlim( (0, fft_delta.size - 1) )
xlabel( "Component" )
ylabel( "Magnitude" )
title( "Magnitude of the Frequency Representation of the Vector delta" )
show()
```

## Prescaling by $\frac{1}{2\pi}$

[numpy has FFT defined such that](http://docs.scipy.org/doc/numpy/reference/routines.fft.html#implementation-details):

$A_k = \sum\limits_{m=0}^{n-1} a_m \exp\left\lbrace -2\pi i \frac{mk}{n} \right\rbrace$

So for the following, we will prescale by $2\pi$.

```python
fft_a = np.fft.fft( a/(2*np.pi) )

# Plotting the magnitude of the vector in frequency domain
figure()
plot( arange( fft_a.size ), abs( fft_a ) )
xlim( (0, fft_a.size - 1) )
xlabel( "Component" )
ylabel( "Magnitude" )
title( "Magnitude of the Frequency Representation of the Vector" )
show()
```

Exponentiating with this and then rescaling:

```python
fft_a2 = fft_a ** 2
a2 = real( np.fft.ifft( fft_a2 ) * 2 * np.pi )

print "||a||   = % 5.3f" % vec_magnitude( a )
print "||a^2|| = % 5.3f" % vec_magnitude( a2 )
```

## Repeated Exponentiation both with and without Scaling

```python
def exp_w_scale( v, n, scale = 2*np.pi ):
    """Exponentiates v to power n with the given scaling value."""
    # Get the Fourier transform of v
    fft_v = np.fft.fft( v / float( scale ) )
    
    # Exponentiate
    fft_vn = fft_v ** n
    
    # Transform back and scale
    vn = real( np.fft.ifft( fft_vn ) ) * float( scale )
    return vn

# Illustrate this for various values of n
ns = arange( 1, 5 )
es1 = map( lambda n : vec_magnitude( exp_w_scale( a, n, 1. ) ), ns )
es2pi = map( lambda n : vec_magnitude( exp_w_scale( a, n ) ), ns )

# Plot the result
figure()
title( "Magnitude of Exponentiated Vectors for Different Scaling Values" )
xlabel( "n" )
ylabel( "||a^n||" )
plot( ns, es1, label="w/o scaling" )
plot( ns, es2pi, label="w/ scaling" )
legend()
show()
```
